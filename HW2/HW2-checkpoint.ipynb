(b) Output all entries that are at least 0.6.
(c) Square all entries of the array.
(d) Reshape it to have 5 rows and 7 columns. (We follow the convention that the first

```python
import numpy as np
# For (a)
A = np.random.uniform(0, 1, 35)
print(A)

# For (b)
point_six_up = A[A >= 0.6]
print(point_six_up)

# For (c)
A = A**2

# For (d)
B = A.reshape(5, 7)
print(B)

# For (e)
median = np.median(B, axis=1)
print("\n The median of each row in B is", median)

# For (f)
mean = np.mean(B, axis = 0)
print("\n The mean of each column in B is", mean)

# For (g) 
np.put_along_axis(B, np.ones((5,1), dtype=np.int16), 0, axis=1)
print("\n Set the second column of B to be all 0: \n",B)

# For (h)
np.put_along_axis(B, np.arange(1, 4).reshape(3, 1), 1, axis=0)
print("\n Set rows from the second to fourth to be all filled with 1: \n",B)



```

## Feature Selection 

```python
# select features 
features = ['sex', 'age', 'famrel', 'freetime', 'goout', 'health', 'absences', 'studytime', 'schoolsup']
```

## Additional Data Preparation

Since we will use the features `sex` and `schoolsup` in the later training and predicting and the `scikit-learn` algorithm only works for numerical attributes, we have to replace the `"Male"/"Female"` and `"yes"/"no"` with `1/0`.  

```python
data['sex'] = data['sex'].apply(lambda x : 0 if x == "F" else 1)
data['schoolsup'] = data['schoolsup'].apply(lambda x : 1 if x == "yes" else 0)
data.dtypes
```

## Data Partition

```python
from sklearn.model_selection import train_test_split 
import numpy as np

X,y = data[features], data['G3']

# Partition the data set
X_train,X_test,y_train,y_test = train_test_split(X, y)
print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape) 

y_train = np.asarray(y_train)
```

```python
from scipy.cluster.vq import whiten

# Normalize the selected features 
X_train = whiten(X_train)
X_train

```

## Create and Train the Algorithm

```python
# setting up the ploting environment
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

fig,ax = plt.subplots()
ax.set_title('k vs. MSE');
ax.set_xlabel('k')
ax.set_ylabel('MSE')

# Create and Train the Algorithm
from scipy.cluster.vq import kmeans2, vq
clusters_MSE = []

for k in range(1, 21):
    print("when k =", k, ": ")

    # Classify the training set into k clusters
    codebook,cluster = kmeans2(X_train,k,minit='++')
    
    # find the mean of y values for each cluster
    cluster_y = []
    
    for i in range(k):
        total_y = []
        index = 0
        for c in cluster: 
            if c == i: 
                total_y.append(y_train[index])
            index +=1 
        mean_y = np.asarray(total_y).mean() 
        #print("mean of y for cluster", i, "is:", mean_y)
  
        cluster_y.append(mean_y) 
    print("cluster_y: ", cluster_y)
    
    
    # assign closest center to each data in the testing set
    encoding,distances = vq(X_test,codebook)
    prediction = [cluster_y[cl] for cl in encoding]    
    
    # Calculate for the accuracy
    # Estimate accuracy using MSE
    difference_squared = 0
    for i in range(len(y_test)):
        difference_squared += (prediction[i] - y_test.iloc[i])**2
    MSE = difference_squared / len(y_test)
    clusters_MSE.append(MSE)
    
    print("when number of k =", k, ", MSE is", MSE, "\n")
    
print("The MSE for all k's are:", clusters_MSE)
# Visualize the outcome
x = np.arange(1,21)
y = np.array(clusters_MSE)

xplot = np.linspace(1,20,num=200,endpoint=True)
plt.plot(x,y,'o')

h = interp1d(x,y,kind='quadratic')
plt.plot(xplot,h(xplot),'-.')
```
